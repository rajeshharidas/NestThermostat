```python
from cassandra.cluster import Cluster
cluster = Cluster(['127.0.0.1'])
session = cluster.connect('mynestthermostat')
```

```python
rows = session.execute('select * from nesteventdata limit 50;')
for row in rows:
    print (row.eventid)
    print (row.timeofevent)
    print (row.traitkey)
    print (row.traitvalue)
    print ('\n')
        
```
